# Creating Flow Diagrams for Projects

```
from graphviz import Digraph
```

```
# Initiate a new Graph Object
graph = Digraph('Breast-Cancer-Prediction-Flowchart', format='png', )
```

```
# add nodes
graph.node('A', 'Add Breast Cancer Dataset in Workspace')
graph.node('B', 'Train Model Using Automated ML')
graph.node('C', 'Get Best Model from Automated ML Run')
graph.node('D', 'Perform Hyperparameter Tuning Using HyperDrive')
graph.node('E', 'Get Best Model from HyperDrive Run')
graph.node('F', 'Compare Both Model and Deploy the best one')
graph.node('G', 'Perform Inference Request on Deployed Endpoint')
```

```
graph.edges(['AB', 'AD', 'BC', 'DE', 'EF', 'CF', 'FG'])
```
